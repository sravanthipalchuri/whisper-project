```python
from datasets import load_dataset

dataset = load_dataset(
    "openslr/librispeech_asr",
    "clean",
    split="test[:20]"
)
```

```python
for sample in dataset:
    prediction = pipe(sample["audio"])
```

```python
predictions=[]
```

```python
for sample in dataset:
    result = pipe(sample["audio"])

    predictions.append({
        "Ground Truth": sample["text"],
        "Prediction": result["text"]
    })
```

```python
import pandas as pd

df = pd.DataFrame(predictions)
```

```python
df.head()
```

```python
df.to_csv("predictions.csv", index=False)
```

```python
from jiwer import Compose, ToLowerCase, RemovePunctuation, Strip, wer

# Normalize both ground truth and prediction
transformation = Compose([
    ToLowerCase(),
    RemovePunctuation(),
    Strip()
])

ground_truth = [
    transformation(text) for text in df["Ground Truth"]
]

predictions = [
    transformation(text) for text in df["Prediction"]
]

score = wer(ground_truth, predictions)

print("Overall WER:", score)
```
